```python
#-----------Excessive sleepiness and fatigue during the day

sleepday_merged['SleepDay'] = pd.to_datetime(sleepday_merged['SleepDay'])

sleepday_merged['SleepHours'] = sleepday_merged['TotalMinutesAsleep'] / 60

sleepday_merged['TimeInBedHours'] = sleepday_merged['TotalTimeInBed'] / 60

sleepday_merged['SleepEfficiency'] = sleepday_merged['SleepHours'] / sleepday_merged['TimeInBedHours']
fatigue_users = sleepday_merged[
    (sleepday_merged['SleepHours'] < 7) &
    (sleepday_merged['SleepEfficiency'] < 0.85)
]

fatigue_users[['Id','SleepHours','SleepEfficiency']].head()

#---------------------A substantial portion of users record less than 6 hours of sleep, indicating a high risk of excessive daytime sleepiness and fatigue.



```

```python
dailyActivity_merged['ActivityDate'] = pd.to_datetime(dailyActivity_merged['ActivityDate'])
dailycalories_merged['ActivityDay'] = pd.to_datetime(dailycalories_merged['ActivityDay'])
dailysteps_merged['ActivityDay'] = pd.to_datetime(dailysteps_merged['ActivityDay'])
```

```python
merged_df = (
     dailyActivity_merged
    .merge(dailycalories_merged, left_on=['Id','ActivityDate'], right_on=['Id','ActivityDay'], how='left')
    .merge(dailysteps_merged, on=['Id','ActivityDay'], how='left')
)
merged_df.head()
```

```python
#-------------1. First use the daily activity data and find the unique id of all the users.

merged_df['Id'].unique()

#--------------The dataset contains multiple unique users, enabling analysis of diverse activity patterns across individuals
```

```python
#---------------2. Then find the number of days they use their fitness tracker.
user_days = merged_df.groupby('Id')['ActivityDate'].nunique().reset_index()
user_days.columns=['deff','dayss']
user_days.head()

#--------------------Users show varying levels of engagement, with the number of days they use the fitness tracker differing significantly across individuals.

```

```python
#--------------3.def user_type(days):

user_days = merged_df.groupby('Id')['ActivityDate'].nunique().reset_index()
user_days.columns = ['Id','dayss']

def user_type(days):
    if days > 20:
        return 'Active'
    elif days >= 10:
        return 'Moderate'
    else:
        return 'Light'

user_category = user_days['dayss'].apply(user_type)
user_category.value_counts()

#-----------------Users can be segmented into Active, Moderate, and Light groups based on the number of days they use their fitness tracker.
```

```python
#----------------4. After that find out the mean distance traveled by each user
mean_distance = merged_df.groupby('Id')['TotalDistance'].mean()
mean_distance.head

#-------------------The average distance traveled varies across users, reflecting differences in daily activity levels and fitness routines.
```

```python
#------------5.. Use this mean distance to group each user into Pro,Intermediate and


def fitness_level(dist):
    if dist > 10:
        return 'Pro'
    elif dist >= 5:
        return 'Intermediate'
    else:
        return 'Beginner'

fitness_category = mean_distance.apply(fitness_level)
fitness_category.value_counts()

#-------------------Users can be effectively classified into Beginner, Intermediate, and Pro fitness levels based on their average distance traveled.
```

```python
#---------------6.Total steps per user
total_steps = merged_df.groupby('Id')['TotalSteps'].sum()
total_steps.head()
#-------------------Total steps vary widely across users, highlighting significant differences in overall physical activity levels
```
